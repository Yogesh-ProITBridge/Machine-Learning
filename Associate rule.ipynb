## 4. Data Preparation

```python
france_retail_data = retail_data[retail_data['Country']=='France']
```

```python
france_retail_data
```

```python
france_data_pivoted=pd.pivot_table(data= france_retail_data,values='UnitPrice',index='InvoiceNo',columns='Description').fillna(0)
```

```python
france_data_one_hot= france_data_pivoted.applymap(lambda x: 1 if x>0 else 0)
france_data_one_hot
```

## 5. Data Mining

```python
!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
#Apriori=frequent itemssets
#Association rules - To find the best associated product
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
france_apriori=apriori(df=france_data_one_hot,
    min_support=0.02,
    use_colnames=True,)
france_apriori
```

```python
best_associated=association_rules(df=france_apriori,
                                 metric='support',
                                 min_threshold=0.02)
```

```python
best_associated
```

```python
best_associated.to_csv()
```
