## 1. Import necessary library

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

## 2. Read the data

```python
Movie_df=pd.read_csv("Movies_dataset.csv")
Movie_df
```

## 3. Data Understanding

```python
tfid=TfidfVectorizer()
```

```python
vectors=tfid.fit_transform(Movie_df['genre'])
```

```python
vectors[0].toarray()
```

```python
sim_dist=linear_kernel(vectors[0],vectors)
sim_dist
```

```python
sim_scores_df=pd.DataFrame(sim_dist).T
sim_scores_df.cloumns=['Scores']
```

```python
sim_scores_df
```

```python
index=pd.Series(data=Movie_df.index,
               index=Movie_df["Movie Title"])
```

```python
def retrival_fun (name,n):
    sim_dist=linear_kernel(vectors[index[name]],vectors)
    sim_scores_df=pd.DataFrame(sim_dist).T
    sim_scores_df.columns=['Scores']
    sim_scores_df=sim_scores_df.sort_values("Scores",ascending=False)

    list=[]

    for i in range (0,n+1):
        result={'Title': Movie_df["Movie Title"][sim_scores_df.index[i]],
                'Rating': Movie_df["Ratinng"] [sim_scores_df.index[i]],
                'Cast' : Movie_df["act"] [sim_scores_df.index[i]],
                'Genre' : Movie_df["genre"] [sim_scores_df.index[i]]
                
               
        }
        list.append(result)

    return list

```

```python
pd.DataFrame(retrival_fun('Dangal',10))
```
